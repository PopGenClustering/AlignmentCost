```python
#@title upload the input file (containing the memberships)

uploaded = files.upload()

for k,v in uploaded.items():
  f1 = os.path.join("/tmp", k)
  with open(f1, 'wb') as f:
    f.write(v)
  print("You've uploaded the file {}".format(v))
```

```python
#@title the upload alignment file (containing the permutations)

uploaded = files.upload()

for k,v in uploaded.items():
  f2 = os.path.join("/tmp", k)
  with open(f2, 'wb') as f:
    f.write(v)
  print("You've uploaded the file {}".format(v))
```

```python
parser = argparse.ArgumentParser()
parser.add_argument('--input_file', type=str, required=True, help="path to membership replicates file")
parser.add_argument('--R', type=int, required=True, help="number of replicates")
parser.add_argument('--perm_file', type=str, required=True, help="path to permutation file")
parser.add_argument('--output_path', type=str, required=False, help="path to save outputs")

input_file = f1 
R = 10 #@param {type:"raw"}
perm_file = f2
output_path = '/tmp/output'

args = parser.parse_args(['--input_file', input_file, 
                          '--R', str(R), 
                          '--perm_file', perm_file,
                          '--output_path', output_path])
```

```python
#@title run program
alignment_cost.main(args)
```

```python
#@title download results
shutil.make_archive(output_path, 'zip', output_path)
files.download(output_path+".zip")
```
